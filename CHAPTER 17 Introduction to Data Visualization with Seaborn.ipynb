## Plotting Statistical Relationships

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.relplot(x = 'sepal_length',
            y = 'sepal_width',
            data = iris)
plt.grid(True)
plt.show()
```

```python
sns.relplot(x = 'sepal_length',
            y = 'sepal_width',
                 data = iris,
                 kind = 'scatter')
plt.grid('on')
plt.show()
```

The function replot() is a generic function where you can pass an argument
to specify the type of plot. You can also create a scatter plot with the function
scatterplot(). For example, the following code creates the same result as shown 

```python
sns.scatterplot(x = 'sepal_width',
                y = 'sepal_length',
                data = iris)
plt.grid('on')
plt.show()
```

```python
sns.relplot(x = 'petal_length',
            y = 'petal_width',
            data = iris)
plt.grid(True)
plt.show()
```

```python
iris
```

```python
sns.scatterplot(x = 'petal_length',
                y = 'petal_width',
                data = iris,
                hue = 'species')
plt.grid(True)
plt.show()
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width',
            data = iris, hue = 'petal_length',
            style = 'species')
plt.grid('on')
plt.show()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_width',
            data = iris, hue = 'petal_length',
            style = 'species')
plt.grid('on')
plt.show()
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width',
            data = iris, hue = 'species', style = 'species')
plt.grid(True)
plt.show()
```

## Plotting Lines

```python
df = pd.DataFrame(np.random.randn(100, 4),
                  index = pd.date_range('1/1/2022',
                                        periods = 100),
                  columns = list('ABCD'))
df = df.cumsum()
df
```

```python
sns.relplot(x = df.index, y = 'A', kind = 'line', data = df)
plt.xticks(rotation = 45) # Rotate the xlabels
plt.show()
```

```python
sns.lineplot(x=df.index,
 y='A', data=df)
plt.xticks(rotation=45)
plt.show()
```

## Visualizing the Distribution of Data

```python
x = np.random.randn(100)
sns.displot(x)
plt.show()
```

You can also make it explicit that you need a histogram in the output as follows

```python
sns.displot(x, kind = 'hist')
plt.show()
```

A histogram is the default kind of graph. You can also show a Gaussian kernel density
estimation (KDE) as follows

```python
sns.displot(x, kind = 'kde')
plt.show()
```

You can visualize an empirical cumulative distribution function (eCDF) as follows:


```python
sns.displot(x, kind = 'ecdf')
plt.show()
```

```python
sns.displot(x, kind = 'hist', kde = True)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(x = 'total_bill', data = tips, kind = 'hist')
plt.show()
```

```python
sns.displot(x = 'total_bill', data = tips, kind = 'hist', bins = 30)
plt.show()
```

```python
sns.displot(x = 'total_bill', data = tips, kind = 'hist', bins = 30, kde = True)
plt.show()
```

```python
tips
```

```python
sns.displot(x = 'total_bill', data = tips, kind = 'kde', hue = 'size')
plt.show()
```

Up to now, we have used a single variable to show the plot. When you use two
variables for plotting, it is known as a bivariate plot. Here is a simple example:

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips, hue = 'size')
plt.show()
```

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', rug = True)
plt.show()
```

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', rug = 'on', bins = 30, kind = 'hist')
plt.show()
```

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
plt.show()
```

```python
sns.displot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde', row = 'size')
plt.show()
```
