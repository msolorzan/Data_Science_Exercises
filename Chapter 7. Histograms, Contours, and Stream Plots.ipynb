# Histograms, Contours, and Stream Plots

## Histograms

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 5, 1, 2, 4, 4, 2, 5, 4, 3, 1, 2]
n_bins = 5
```

```python
plt.hist(x, bins = n_bins)
plt.show()
```

```python
np.random.seed(31415)
n_points = 10000
n_bins = 15
x = np.random.randn(n_points)
plt.hist(x, bins = n_bins)
plt.show()
```

```python
y = np.random.randn(n_points)
plt.hist2d(x, y, bins = 100) # 100 x 100
plt.show()
```

```python
plt.hist(x, 20, density=True,
 histtype='stepfilled',
 facecolor='g', alpha=0.5)
plt.show()
```

```python
plt.hist(x, 100, density = False,
         histtype = 'stepfilled',
         facecolor = 'g', alpha = 1)
plt.show()
```

```python
plt.hist(x, 100, density = True,
         histtype = 'stepfilled',
         facecolor = 'g', alpha = 1)
plt.show()
```

```python
plt.hist(x, 10, density = True,
         histtype = 'bar',
         facecolor = 'g', alpha = 1)
plt.show()
```

```python
plt.hist(x, 10, density = True,
         histtype = 'barstacked',
         facecolor = 'g', alpha = 1)
plt.show()
```

```python
plt.hist(x, 100, density = True,
         histtype = 'step',
         facecolor = 'g', alpha = 1)
plt.show()
```

```python
x = np.arange(-3, 3, 0.005)
y = np.arange(-3, 3, 0.005)
X, Y = np.meshgrid(x, y)
X, Y
```

```python
Z = (X ** 2 + Y ** 2)
Z
```

```python
out = plt.contour(X, Y, Z)
plt.clabel(out, inline=True,
 fontsize=10)
plt.show()
```

```python
out = plt.contour(X, Y, Z)
plt.clabel(out, inline=False,
 fontsize=10)
plt.show()
```

```python
out = plt.contour(X, Y, Z)
plt.clabel(out, inline=True,
 fontsize=30)
plt.show()
```

```python
help(plt.clabel)
```

```python
help(plt.contour)
```

```python
out = plt.contour(X, Y, Z)
plt.clabel(out, inline = True,
           fontsize = 10)
plt.colorbar()
plt.show()
```

```python
out = plt.contour(X, Y, Z,
                  colors = 'g',)
plt.clabel(out,
           fontsize = 10,
           inline = True)
plt.show()
```

```python
plt.contourf(X, Y, Z,
             hatches = ['-', '/', '\\', '//'],
             cmap = 'cool',
             alpha = 0.75)
plt.show()
```

```python
plt.contourf(X, Y, Z,
             hatches = ['|', '-', '\\', '*'],
             cmap = 'cool',
             alpha = 0.75)
plt.show()
```

## Visualizing Vectors with Stream Plots

```python
Y , X = np.mgrid[-5:5:200j, -5:5:300j]
Y, X
```

```python
len(X), len(Y), X.ndim, X.shape
```

```python
U = X ** 2 + Y ** 2
V =X + Y
```

```python
plt.streamplot(X, Y, U, V)
plt.show()
```

```python
plt.streamplot(X, Y, X, Y)
plt.show()
```

```python
plt.streamplot(X, Y, X, V)
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
               density = [0.5, 0.75])
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
               linewidth = 1, color = V,
               cmap = 'cool')
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
               linewidth = 1, color = V,
               cmap = 'Accent')
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
               linewidth = 1, color = V,
               cmap = 'prism')
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
 density=0.6,
 color='k',
 linewidth=X)
plt.show()
```

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
plt.quiver(X, Y, U, V)
plt.show()
```
