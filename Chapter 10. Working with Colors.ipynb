# Working with Colors

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## colorbar()

```python
N = 100
X, Y = np.meshgrid(np.linspace(-5, 5, N),
                   np.linspace(-5, 5, N))
Z = (X ** 2 + Y ** 2)
img = plt.imshow(Z, cmap = 'YlGnBu_r')
plt.show()
```

```python
img = plt.imshow(Z, cmap = 'YlGnBu_r')
plt.colorbar(img)
plt.show()
```

```python
img = plt.imshow(Z, cmap = 'YlGnBu_r')
plt.colorbar(img, shrink = 0.5)
plt.show()
```

```python
img = plt.imshow(Z, cmap = 'coolwarm')
plt.colorbar(img, location = 'bottom', shrink = 0.3)
plt.show()
```

```python
img = plt.imshow(Z, cmap = 'coolwarm')
plt.colorbar(img, extend = 'both')
plt.show()
```
