````python
def sinusoidal(style: str):
    plt.style.use(style)
    fig, ax = plt.subplots()
    
    ax.plot(data, np.sin(data), label = 'Sine')
    ax.plot(data, np.cos(data), label = 'Cosine')
    ax.legend()
````

How to use constrained-layout to fit plots within your figure cleanly.

constrained_layout automatically adjusts subplots and decorations like legends and colorbars so that they fit in the figure window while still preserving, as best they can, the logical layout requested by the user.

constrained_layout is similar to tight_layout, but uses a constraint solver to determine the size of axes that allows them to fit.

constrained_layout typically needs to be activated before any axes are added to a figure. Two ways of doing so are

````python
import matplotlib.pyplot
import matplotlib.gridspec as gridspec

fig = matplotlib.pyplot.figure(constrained_layout=True)
specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
ax1 = fig.add_subplot(specs[0, 0])
ax2 = fig.add_subplot(specs[0, 1])
ax3 = fig.add_subplot(specs[1, 0])
ax4 = fig.add_subplot(specs[1, 1])
plt.show()
````

````python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')
ax2 = fig.add_subplot(gs[1, :])
ax2.set_title('gs[1, :]')
ax3 = fig.add_subplot(gs[2, :])
ax3.set_title('gs[2, :]')
plt.show()
````

````python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[:, 0])
ax1.set_title('gs[:, 0]')
ax2 = fig.add_subplot(gs[:, 1])
ax2.set_title('gs[:, 1]')
ax3 = fig.add_subplot(gs[:, 2])
ax3.set_title('gs[:, 2]')
plt.show()
````

````python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3) # Size = 3, 3
ax1 = fig.add_subplot(gs[0, :]) # row = 0, columns = all
ax1.set_title('gs[0, :]')
ax2 = fig.add_subplot(gs[1, :-1]) # row = 1, columns = all - 1
ax2.set_title('gs[1, :-1]')
ax3 = fig.add_subplot(gs[1:, -1]) # 
ax3.set_title('gs[1:, -1]')
ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')
ax5 = fig.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')
plt.show()
````

## Lines, Bars, and Scatter Plots

````python
data = np.linspace(0, 9, 10)
plt.plot(data)
plt.show()
````

````python
t = np.arange(0.01, 10, 0.01)
plt.semilogx(t, np.cos(2 * np.pi * t))
plt.show()
````

````python
plt.plot(t, np.cos(2 * np.pi * t))
plt.show()
````

````python
plt.semilogy(t, np.cos(2 * np.pi * t))
plt.show()
````

````python
plt.loglog(t, np.cos(2 * np.pi * t))
plt.show()
````

````python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
ye = np.random.rand(len(x))/10
plt.errorbar(x, y, yerr = ye)
plt.show()
````

````python
xe = np.random.rand(len(x))/10
plt.errorbar(x, y, xerr = xe)
plt.show()
````

````python
plt.errorbar(x, y, xerr = xe, yerr = ye)
plt.show()
````

## Bar Graphs

````python
x = np.arange(4)
y = np.random.rand(4)
plt.bar(x, y)
plt.show()
````

````python
x, y
````

````python
y = np.random.rand(3, 4)
plt.bar(x, y[0], color = 'b', width = 0.25)
plt.bar(x + 0.25, y[1], color = 'g', width = 0.25)
plt.bar(x + 0.5, y[2], color = 'r', width = 0.25)
plt.show()
````

````python
x = np.arange(4)
y = np.random.rand(4)
plt.barh(x, y)
plt.show()
````

````python
y = np.random.rand(3, 4)
plt.barh(x, y[0], color = 'b', height = 0.25)
plt.barh(x + 0.25, y[1], color = 'g', height = 0.25)
plt.barh(x + 0.5, y[2], color = 'r', height = 0.25)
plt.show()
````

## Scatter Plot

````python
N = 1000
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
size = (20)
plt.scatter(x, y, c = colors, s = size, alpha = 1)
plt.show()
````
